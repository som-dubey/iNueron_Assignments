Q30. Write a pandas program to get current date, oldest date and
number of days between Current date and the oldest date of Ufo
dataset.

```python
import datetime
from datetime import date
ufo = pd.read_csv('ufo_sighting_data.csv')
ufo['datetime'] = pd.to_datetime(ufo.datetime, errors='coerce').dt.strftime('%d-%m-%Y')

ufo = ufo.sort_values(by = 'datetime')
oldest_date = ufo.datetime.head()
oldest_date = oldest_date.iloc[0]
current_date = date.today().strftime("%d-%m-%Y")  
def question_30(d1, d2):
    d1 = datetime.datetime.strptime(d1, "%d-%m-%Y")
    d2 = datetime.datetime.strptime(d2, "%d-%m-%Y")
    return abs((d2 - d1).days)
```

```python
days_between = question_30(oldest_date,current_date)
days_between
```

Q34. Write a pandas program to get the difference (in days) between
documented date and reporting date of unidentified flying object
(UFO).

```python
from datetime import date
ufo = pd.read_csv('ufo_sighting_data.csv')
ufo = ufo.sort_values(by = 'datetime')
ufo['datetime'] = pd.to_datetime(ufo.datetime, errors='coerce', infer_datetime_format=True).dt.date
ufo['date posted'] = pd.to_datetime(ufo['date posted'], errors='coerce',infer_datetime_format=True).dt.date
ufo = ufo.dropna(subset = ['date posted', 'datetime'] )
ufo['Days_diff'] = ((ufo['date posted'] - ufo['datetime'])/ np.timedelta64(1, 'D')).round().astype(int)
ufo.head()
```

Q40. Write a pandas program to create a Timewheel of Hour VS Year
comparison of the top 10 years in which the (UFO) was sighted

```python
from datetime import date
ufo = pd.read_csv('ufo_sighting_data.csv')
ufo = ufo.sort_values(by = 'datetime')
ufo['datetime_year'] = pd.to_datetime(ufo.datetime, errors='coerce', infer_datetime_format=True).dt.year
ufo['datetime_hour'] = pd.to_datetime(ufo.datetime, errors='coerce', infer_datetime_format=True).dt.hour

ufo['Year_hour_count'] = ufo.groupby(['datetime_year', 'datetime_hour'])['datetime'].transform('count')
ufo = ufo.sort_values(by = 'Year_hour_count', ascending = False).dropna(subset = ['Year_hour_count'])
ufo_1 = ufo.loc[:, ['datetime_year', 'datetime_hour', 'Year_hour_count']].drop_duplicates()\
.sort_values(by = 'Year_hour_count', ascending = False).reset_index(drop = True)
ufo_1 = ufo_1.head(10)
```

```python
import matplotlib as mpl
import matplotlib.cm as cm

def pie_heatmap(ufo_1, cmap='coolwarm_r', vmin=None, vmax=None,inner_r=0.25, pie_args={}):
    n, m = ufo_1.shape
    vmin= ufo_1.min().min() if vmin is None else vmin
    vmax= ufo_1.max().max() if vmax is None else vmax
    
    centre_circle = plt.Circle((0,0),inner_r,edgecolor='black',facecolor='white',fill=True,linewidth=0.25)
    
    plt.gcf().gca().add_artist(centre_circle)
    norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)
    
    cmapper = cm.ScalarMappable(norm=norm, cmap=cmap)
    
    for i, (row_name, row) in enumerate(ufo_1.iterrows()):
        
        labels = None if i > 0 else ufo_1.columns
        wedges = plt.pie([1] * m,radius=inner_r+float(n-i)/n, colors=[cmapper.to_rgba(x) for x in row.values],
                         labels=labels, startangle=90, counterclock=False, wedgeprops={'linewidth':-1}, **pie_args)
        plt.setp(wedges[0], edgecolor='grey',linewidth=1.5)
        wedges = plt.pie([1], radius=inner_r+float(n-i-1)/n, colors=['w'], labels=[row_name], startangle=-90, wedgeprops={'linewidth':0})
        plt.setp(wedges[0], edgecolor='grey',linewidth=1.5)
plt.figure(figsize=(8,8))
plt.title("Timewheel of Hour Vs Year",y=1.08,fontsize=30)
pie_heatmap(ufo_1, vmin=-20,vmax=80,inner_r=0.2)
```
